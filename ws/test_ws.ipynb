```python
"""
1. На языке Python реализовать алгоритм (функцию) определения четности целого числа, 
который будет аналогичен нижеприведенному по функциональности, но отличен по своей сути. 
Объяснить плюсы и минусы обеих реализаций.
Python example:
def isEven(value):return value%2==0
"""

def is_even(num):
    """
    + не выполняет операцию деления 
    """
    return int(bin(num)[-1]) == 0
    
def isEven(value):
    """
    + понятность реализации
    """
    return value%2==0
    
for i in range(-100,100):
    assert is_even(i) == isEven(i)    
```

```python
"""
2. На языке Python (2.7) реализовать минимум по 2 класса реализовывающих циклический буфер FIFO. 
Объяснить плюсы и минусы каждой реализации.
"""

def check_fifo(FifoCalss):
    check_list = [i for i in range(21)] 
    print(f"\n{FifoCalss}")
    fifo = FifoCalss()

    assert fifo.getData() == None
    
    for val in check_list:          
        fifo.addData(val)
        
    result = []    
    
    for i in range(len(check_list)):
        result.append(fifo.getData())
        
    print(check_list)        
    print(result)    
    assert check_list == result
    
    if FifoCalss.__name__ == 'FIFO_3':
        assert fifo.getPointers() == (0, 0)
        for val in check_list:          
            fifo.addData(val)
        
        result = []    
    
        for i in range(len(check_list) // 2):
            result.append(fifo.getData())
            
        fifo.reset_index()
        data = fifo.getAllData() 
        assert fifo.getPointers() == (0, len(data))
        
        
        
            

#1.==============================================
class FIFO_1 :
    """
    Класс fifo на списках, хранит индекс следующего элемента в очереди.
    + требует минимального времени на добавления данных и получение данных
    - не освобождает память
    , но может быть актуальным при необходимости хранит историю.
    """    

    
    def is_empty_data(self):
        return len(self.data) == 0
    
    def __init__(self):         
        self.next = 0
        self.data = []

        
    def addData(self, new_data):        
        self.data.append(new_data)
     
    
    def getData(self):             
        if self.is_empty_data() > 0:
            return None
        result = self.data[self.next]
        self.next += 1
        return result
    
    
check_fifo(FIFO_1)


#2.==============================================    
class FIFO_2(FIFO_1) :         
    """
    Класс fifo на списках, удаляет из списка первый элемент    
    + требует минимального времени на добавление данных
    + освобождает память, 
    - но при больших объёмах данных будет требовать много времени на пересчёт индексов O(n)
    """ 
    
    
    def getData(self):            
        if self.is_empty_data() > 0:
            return None
        result = self.data[self.next]
        del self.data[self.next]
        return result        
        

check_fifo(FIFO_2)    
        
    
#3.==============================================     
class FIFO_3(FIFO_1) :
    """
    Класс fifo на словарях, хранит числовой ключ id нового и следующего элемента в очереди.
    + требует минимального времени на добавление и извлечение данных
    + освобождает память
    - dict занимает больше памяти чем list
    - при долгой работе значения указателей индексов next и new могут значительно вырасти и их придётся сбрасывать
    """    
    def __init__(self):         
        self.next, self.new = 0, 0
        self.data = {}

        
    def addData(self, new_data):         
        self.data[self.new] = new_data
        self.new += 1        
        
     
    def getData(self):        
        if self.is_empty_data():
            self.__init__()
            return None
        result = self.data[self.next]
        del self.data[self.next]        
        if self.is_empty_data():
            self.__init__()
        else:
            self.next += 1
        return result    
    
    
    def reset_index(self):
        for i in range(len(self.data)):
            self.data[i] = self.data[i+self.next]
            del self.data[i+self.next]
        self.new = i+1
        self.next = 0
        
        
    def getPointers(self):
        return self.next, self.new
    
    
    def getAllData(self):
        return self.data

    
check_fifo(FIFO_3)     
```

```python
"""
3. На языке Python реализовать функцию, которая быстрее всего (по процессорным тикам) 
отсортирует данный ей массив чисел. Массив может быть любого размера со случайным порядком чисел 
(в том числе и отсортированным). Объяснить почему вы считаете, что функция соответствует заданным критериям.
"""


#Алгортьи сортировка слиянием работант за ыремя O(nlogn)

def check_sort(arr, f):    
    mrg_sort_arr = f(arr.copy())
    print(f"\n{f.__name__}:")
    print(arr)
    arr.sort()
    print(mrg_sort_arr)
    assert arr == mrg_sort_arr

def mrg(a, b):
    c = []
    while len(a) > 0 and len(b) > 0:        
        if a[0] > b[0]:
            c.append(b[0])
            del b[0]
        else:
            c.append(a[0])
            del a[0]
    c.extend(a)
    c.extend(b)
    return c


def sort_merge(arr):    
    for i in range(len(arr)): 
        arr[i] = [arr[i]]
    cnt = 0
    while len(arr) > 1:
        if cnt+1 >= len(arr):
            cnt = 0
        arr[cnt] = mrg(arr[cnt],arr[cnt+1])
        del arr[cnt+1]
        cnt += 1
    return arr[0]
        
arr_1 = [5,8,4,0,64,568,9,4,0,0,5,69,37]        
check_sort(arr_1, sort_merge)

"""
Если массив содержит относительно не большие повторяющинся символы 
то отсортировать его можно за время O(n) алгоритмом сортировки подсчётом
"""
M = 10
def sort_count(A):
    B = [0 for i in range(0,M+1)]
    Ast = [0 for i in range(0,len(A))]
    for i in range(len(A)):
        B[A[i]] += 1
    for i in range(2,M+1):
        B[i] += B[i-1]
    for j in range(len(A)-1,-1,-1):
        Ast[B[A[j]]-1] = A[j]
        B[A[j]] = B[A[j]] -1
    return Ast


arr_2= [2, 3, 9, 2, 9, 7, 4, 8, 1, 9]
check_sort(arr_2, sort_count)
```
